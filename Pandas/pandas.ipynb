# advanced example on index you can Skip it

```python
data = {
  "age": [50, 40, 30, 40, 20, 10, 30],
  "qualified": [True, False, False, False, False, True, True]
}

df = pd.DataFrame(data)
# by defulat the index is num from 0 
print(df)
```

```python
data = {
  "age": [50, 40, 30, 40, 20, 10, 30],
  "qualified": [True, False, False, False, False, True, True]
}

index = ["Mary", "Sally", "Emil", "Tobias", "Linus", "John", "Peter"]

df = pd.DataFrame(data, index = index)
# the index become String from the list
newdf = df.sort_index()

print(newdf)
```
