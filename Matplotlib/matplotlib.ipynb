```python
x=['c++','java','python','c','c#','php']
votes=[5,6,7,8,9,10]
plt.pie(votes,labels=x,explode=[0.1,0.2,0,0,0,0],autopct='%.2f%%')
plt.legend()
 # pie chart --> f float (%.2f) --> 2 number after decimal
# auto pct --> percentage of votes on pie
# i integer %i%%
```

# Histogram
A histogram is a graph showing frequency distributions.

It is a graph showing the number of observations within each given interval


2 people from 140 to 145cm

5 people from 145 to 150cm



np.random.normal(Mean,std,1000) --> 1000 samples from normal distribution with mean = 0 and std = 1

```python
ages=np.random.normal(20,1.5,200)
# plt.hist(ages,bins=5) # histogram --> bins = number of bars
plt.hist(ages,bins=[ages.min(),18,21,ages.max()],color='red')
# bins = [ages.min(),18,21,ages.max()] --> values from min to max 
plt.show()
```

```python
plt.boxplot(ages,)
```
