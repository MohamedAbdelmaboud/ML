```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset('tips')
iris=sns.load_dataset('iris')
titanic=sns.load_dataset('titanic')
planets=sns.load_dataset('planets')
anagrams=sns.load_dataset('anagrams')
```

```python
tips
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='smoker',size='size',palette='coolwarm')
# size of the marker depending on the size of the bill
# hue --> different colors for different groups of data somker or not 
# sns.scatterplot(x='total_bill',y='tip',data=tips,hue='day')
# hue --> different colors for different groups of data based on day
```

```python
sns.histplot(tips['total_bill'] , kde=True, bins=15 ),
# kde --> kernel density estimation 
# (KDE) plot along with the histogram.
#  A KDE plot is a smooth curve
```

```python
# sns.barplot(x='sex',y='tip',data=tips,palette='coolwarm')
sns.barplot(x=tips['sex'],y=tips['tip'],palette='coolwarm')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette='rainbow',hue='smoker')
```

```python
sns.stripplot(x='day',y='tip',data=tips,palette='coolwarm',hue='sex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde',fill=True)
```

Correlation is a key statistical concept that researchers employ to analyze connections within their data. It helps us to Understand the Relationship Between Variables

```python
sns.pairplot(titanic.select_dtypes(include='number'),palette='coolwarm')
```

```python
titanic.corr()
```

```python
sns.heatmap(titanic.corr(),cmap='coolwarm',annot=True)
```
